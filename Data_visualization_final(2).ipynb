```python
points = pd.read_csv("./datas/movie_points.csv", index_col=[0])
points.head(15)
```

```python
col = ['Sales', 'Point', 'Audience', 'Screen']
points_norm = round(points[col] / points[col].max(), 4)
points10 = points_norm.head(15)
points10['Movie'] = points['Movie'].head(15)
points10.set_index('Movie', inplace=True)
points10
```

```python
labels = list(points['Movie'].head(10))
```

```python
from matplotlib import font_manager
f_path = "C:\Windows\Fonts\malgun.ttf"
font_manager.FontProperties(fname=f_path).get_name()
from matplotlib import rc
rc('font', family='Malgun Gothic')
%matplotlib inline


sns.set_style('darkgrid')
plt.figure(figsize=(10,6))
plt.title('Sales, audience, screen, points for TOP15 movies')
plt.xlabel('Movie')
plt.plot(points10['Sales'], '-b', label='sales')
plt.plot(points10['Audience'], '-g', label='audience')
plt.plot(points10['Screen'], '-r', label='screen')
plt.plot(points10['Point'], '-y', label='points')
plt.xticks(fontsize=10, rotation=45, fontproperties=font_manager.FontProperties(fname=f_path).get_name())
plt.legend(loc='best');
```
